```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Establishing a connection to the MySQL database
connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="qwerty1234",
    database="finance_database"
)

# Defining the SQL query and reading the results into DataFrame
query = "SELECT * FROM TABLE_NAME"
df = pd.read_sql_query(query, connection)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Plotting the count of genders
plt.figure(figsize=(5,3))
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution', fontsize=14)
plt.xlabel('Gender', fontsize=12)  
plt.ylabel('Count', fontsize=12)  
plt.show()
```

```python
# Plotting a histogram of ages
sns.histplot(data=df, x='age', bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting a box plot of investment avenues
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']])
plt.title('Investment Avenues', fontsize=16)
plt.xlabel('Avenue', fontsize=14)
plt.ylabel('Amount', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Duration Analysis
duration_counts = df["Duration"].value_counts()
duration_counts.plot(kind="bar", rot=0)
plt.xlabel("Duration")
plt.ylabel("Count")
plt.title("Preferred Investment Durations")
plt.show()
```

```python
# Monitoring Frequency using bar graph
plt.figure(figsize=(6,3))
monitoring_counts = df["Invest_Monitor"].value_counts()
monitoring_counts.plot(kind="bar", rot=0)
plt.xlabel("Monitoring Frequency")
plt.ylabel("Count")
plt.title("Investment Monitoring Frequency")
plt.show()
```
